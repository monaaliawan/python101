```python
from google.colab import drive
drive.mount('/content/drive')

# Your code to read the target file (10pt)


f_folder = "/content/drive/MyDrive/Python Programming"
f_name = "/country_import_export_2020.csv"
f_path = f_folder + f_name
print(f_path)

with open(f_path, 'r') as f:
  print(f.read())

# Your code to create two dictionaries as follows (20pt):
# import_dict = {"France": 520562.93, "Germany": 1348649.76 ....}
# export_dict = {"France": 600915.45, "Germany": 1167605.45 ....}
import csv
import_dict = {}
export_dict = {}
with open(f_path, 'r') as f:
    reader=csv.reader(f)
    next(reader)
    for row in reader:
      Country, Import, Export = row
      import_dict[Country] = float(Import)
      export_dict[Country] = float(Export)

print("import Dictionary:", import_dict)
print("export Dictionary:", export_dict)





Okay, now let's check some basic statistics.

- Count the number of countries in the dataset (__5pt__)
- Print the minimum and maximum values for export and import, separately (__10pt__).



# Your code to count the number of countries (5pt)
No_of_countries = len(import_dict)
print("Number of countries:", No_of_countries)

# Your code to print out min & max values for export & import (10pt)
min_export = min(export_dict.values())
max_export = max(export_dict.values())
min_import = min(import_dict.values())
max_import = max(import_dict.values())

print("Minimum export value:", min_export)
print("Maximum export value:", max_export)
print("Minimum import value:", min_import)
print("Maximum import value:", max_import)


Next, we create another dictionary `net_dict`, whose keys are countries (same as `import_dict` and `export_dict`) and values are the net values between export and import (amount of export - amount of import) (__15pt__). Of course, if a country imported more products than exported, the value can be negative.


# Your code to create a new dictionary, net_dict (15pt)
net_dict = {}
for country in import_dict:
  if country in export_dict:
    net_dict[country] = export_dict[country] - import_dict[country]
print(net_dict)

Finally, sort the countries based on the net values in `net_dict` --- which can be calculated by (the amount of export) - (the amount of import) --- in the descending order. This is something we did not cover during our practices, so you need to find out in Google how to sort keys by their values in a dictionary (__25pt__).


# Your code to sort dictionary by values (25pt)
sorted_net_dict = dict(sorted(net_dict.items(), key=lambda item: item[1], reverse=True))
print(sorted_net_dict)

Save your result to a new file with the name `midterm_result_(your school id).txt`(__15pt__).

# Your code to save the result (15pt)
school_id = '202332090'  # Replace with your actual school ID
result_file_path = f"/content/drive/MyDrive/Python Programming_Midterm_Result2024_{school_id}.txt"

with open(result_file_path, 'w') as file:
    for country, net_value in sorted_net_dict.items():
        file.write(f"{country}: {net_value}\n")

print(f"Results saved to {result_file_path}")
```
